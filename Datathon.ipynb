```python
datasets={'Demographics':set(demographics.columns),
          "Geographic":set(geographic.columns),
          "Weather":set(weather.columns),
          "Zones":set(zones.columns),
          "Yellow trips":set(yellow_trips.columns),
          "Green trips":set(green_trips.columns),
          "MTA Trips":set(mta_trips.columns),
          "Uber trips 2014":set(uber_trips_2014.columns),
          "Uber trips 2015":set(uber_trips_2015.columns)}

print("Common columns")
for ds,vds in datasets.items():
    for k,v in datasets.items():
        if ds != k:
            print(f'{ds} - {k} : {vds&v}')
    print('-'*50)
            
```
